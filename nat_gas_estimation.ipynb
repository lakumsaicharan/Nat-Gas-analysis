```python
last_date = data['Dates'].max()
last_date
```

```python
year_start = last_date - pd.DateOffset(months = 11)
year_start
```

```python
pattern = data[data['Dates'] >= year_start]
pattern
```

```python
dates = pd.date_range(start = last_date + pd.DateOffset(months = 1),periods = 12,freq = 'ME')
dates
```

```python
prices = pattern['Prices'].values
prices
```

```python
new_data = pd.DataFrame({'Dates':dates,'Prices':prices})
new_data
```

```python
df = pd.concat([data,new_data]).reset_index(drop = True)
df
```

```python
def estimate(date):
    date = pd.to_datetime(date)
    if date <= data['Dates'].max():
        return np.interp(date.toordinal(),data['Dates'],data['Prices'])
    elif date <= new_data['Dates'].max():
        match = new_data[new_data['Dates']==pd.to_datetime(date)]
        if not match.empty:
            return match['Prices'].values[0]
        else:
            return None
    else:
        return None
```

```python
estimate('2024-10-31')
```
